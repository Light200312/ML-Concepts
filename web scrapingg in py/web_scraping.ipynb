````python
import bs4
import requests
import pandas as pd

print(f"BeautifulSoup version: {bs4.__version__}")
print(f"Requests version: {requests.__version__}")
````

````python
res=requests.get("https://news.ycombinator.com")
soup=bs4.BeautifulSoup(res.text,'lxml')
print(soup.title)
````

````python
articles=soup.find_all('tr',class_='athing')
print(articles[0].find_all('td',class_='title')[1].find('a').text)
print(articles[0].find_all('td',class_='title')[0].find('span').text)
print(articles[0].find_all('td')[2].find_all('span',class_='sitestr')[0].text)
print(articles[0].find_all('td',class_='title')[1].find('a').get('href'))


````

````python
titles=[]
urls=[]
site_names=[]
for i in range(0,30):
    if i==1: print(len(articles[i].find_all('td',class_='title')[1].find('a').text))
    print('\n\n')
  
    title =articles[i].find_all('td',class_='title')[1].find('a').text
    url=articles[i].find_all('td',class_='title')[1].find('a').get('href')
    site_name=articles[i].find_all('td')[2].find_all('span',class_='sitestr')[0].text
    print(title)
    print(url)
    print(site_name)

    titles.append(title)
    urls.append(url)
    site_names.append(site_name)

    
````

````python
print(len(titles))
print(len(urls))
print(len(site_names))
# print(titles)
df=pd.DataFrame({
    "title":titles,
    "url":urls,
    "site_name":site_names
})
df
````

When you hit a "wall of protection" (like Cloudflare, Akamai, or DataDome), standard libraries like `requests` and `BeautifulSoup` fail because they don't behave like a real human browser. They lack headers, cookies, and the ability to execute JavaScript.

Here are the four most effective ways to bypass bot protection:

---

## 1. Use a "Stealth" Browser (Playwright or Selenium)

Most modern protection checks if the browser can run JavaScript. Standard scrapers can't, so they get blocked instantly.
**Playwright with the Stealth plugin** is the gold standard right now. It mimics human behavior like mouse movements and screen resolutions.

```python
# Install: pip install playwright playwright-stealth
from playwright.sync_api import sync_playwright
from playwright_stealth import stealth_sync

with sync_playwright() as p:
    browser = p.chromium.launch(headless=True)
    page = browser.new_page()
    
    # This line makes the bot look like a real user
    stealth_sync(page)
    
    page.goto("https://blocked-site.com")
    print(page.title())
    browser.close()

```

---

## 2. Rotate Your User-Agent

A User-Agent tells the server what browser and OS you are using. If you send 500 requests with the default "python-requests/2.28" header, you will be banned.

**The Fix:** Use a real browser string and rotate it.

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36'
}
response = requests.get(url, headers=headers)

```

---

## 3. Use `cloudscraper` (For Cloudflare only)

Cloudflare is the most common protection (the "Just a moment..." screen). A library called `cloudscraper` is designed specifically to solve these challenges automatically.

```python
# Install: pip install cloudscraper
import cloudscraper

scraper = cloudscraper.create_scraper() # Returns a requests-like object
print(scraper.get("https://website-with-cloudflare.com").text)

```

---

## 4. Use Residential Proxies

If you make too many requests from one IP, that IP gets "blacklisted."

* **Datacenter IPs:** Cheap, but easily detected (AWS, Google Cloud IPs).
* **Residential IPs:** These look like they come from a home Wi-Fi (Verizon, Comcast). They are much harder for websites to block.

---

## Summary of Tactics

| Level of Protection | Recommended Tool |
| --- | --- |
| **Low** (Checks headers) | `requests` + Real User-Agent |
| **Medium** (Cloudflare/JS Check) | `cloudscraper` or `Playwright` |
| **High** (Behavioral Analysis) | `Playwright Stealth` + Residential Proxies |
| **Extreme** (CAPTCHAs) | 2Capcha / Anti-Captcha API Services |

Scraping behind protection often violates a site's **Terms of Service**. Always check `robots.txt` (e.g., `example.com/robots.txt`) to see what the site allows.
